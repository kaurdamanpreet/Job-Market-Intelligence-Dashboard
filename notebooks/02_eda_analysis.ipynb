Load Clean data

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("cleaned_canada_jobs.csv")
df.head()
```

Which roles are most common in the Canadian job market?

```python
job_counts = df["job_role"].value_counts()
job_counts
```

Data Analyst and Data Scientist roles usually dominate

AI / ML Engineer roles show strong demand growth

```python
job_counts.plot(kind="bar", title="Job Demand by Role")
plt.show()
```

alary Analysis by Role (CAD)

Which roles pay the most on average?

```python
avg_salary_role = df.groupby("job_role")["salary_cad"].mean().sort_values(ascending=False)
avg_salary_role
```

Insight:

AI / ML Engineer roles typically have the highest average salary

Entry-level analyst roles have lower but stable compensatio

4Ô∏è‚É£ Entry-Level vs Experienced Salary Comparison
üîç Question:

How much does experience impact salary?

```python
salary_by_exp = df.groupby("experience_level")["salary_cad"].mean()
salary_by_exp
```

Insight:

Clear salary jump from Entry-level ‚Üí Mid-level

Senior roles command significantly higher compensation

This becomes a great Tableau comparison chart.

5Ô∏è‚É£ Job Demand by Experience Level

```python
exp_counts = df["experience_level"].value_counts()
exp_counts
```

Insight:

Entry-level roles exist but are fewer

Highlights competitiveness of junior job market

6Ô∏è‚É£ Trend Analysis Over Time
üîç Question:

Is demand growing over the years?

```python
jobs_by_year = df.groupby("work_year").size()
jobs_by_year
```

```python
jobs_by_year.plot(marker="o", title="Job Postings Trend Over Time (Canada)")
plt.show()
```

Insight:

Data and AI roles show steady growth post-2020

Validates long-term career potential
